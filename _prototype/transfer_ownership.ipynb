```python
from Preamble import *
```

```python
NewAnswers = pd.read_csv('Answers_created.txt', header=[0], delimiter='\t')
NewLiveboards = pd.read_csv('Pinboards_created.txt', header=[0], delimiter=',')
user_map = pd.read_csv('User_names.csv', header=[0], delimiter= ',')
```

```python
AnswersOwner = pd.merge(NewAnswers, user_map, how="left", on=["author"])
LiveboardsOwner = pd.merge(NewLiveboards, user_map, how="left", on=["author"])
```

```python
LiveboardsOwner = LiveboardsOwner[['id','name_x','name_y']]
AnswersOwner = AnswersOwner[['id','name_x','name_y']]
```

```python
print("Backup Year: ")
backup_year = input()
print("Backup Month: ")
backup_month = input()
print("Backup Day: ")
backup_day = input()
import datetime
backupdate = (datetime.datetime(int(backup_year or 0),int(backup_month or 0),int(backup_day or 0),0,0) - datetime.datetime(1970,1,1)).total_seconds()

sess = requests.Session()
loginCloud(sess)
AnswerHeader = listAnswers(sess).json()
try:
    df = json_normalize(AnswerHeader)
    df['date_created'] = pd.to_datetime(df['created']/1000,unit='s')
    df['created_flag'] = df['created'] > backupdate * 1000
    new_answer_objects = df[df['created_flag']==True]
    new_answer_objects = new_answer_objects[['id','name']]
    print("Number of Answers created " ,len(new_answer_objects.index))
except:
    pass
PinboardHeader = listPinboards(sess).json()
try:
    df = json_normalize(PinboardHeader)
    df['date_created'] = pd.to_datetime(df['created']/1000,unit='s')
    df['created_flag'] = df['created'] > backupdate * 1000
    new_liveboard_objects = df[df['created_flag']==True]
    new_liveboard_objects = new_liveboard_objects[['id','name']]
    print("Number of Liveboards created " ,len(new_liveboard_objects.index))
except:
    pass
```

```python
df_answers = pd.merge(AnswersOwner,new_answer_objects, left_on='name_x', right_on='name')
df_liveboards = pd.merge(LiveboardsOwner,new_liveboard_objects, left_on='name_x', right_on='name')
```

# Transfer Ownership for all objects

```python
sess = requests.Session()
loginCloud(sess)
df_answers = df_answers.replace('@','%40', regex=True) # replacing @ character for all users

for i in range(len(df_answers)):
    try:
        ObjGUID = df_answers['id_y'].iloc[i]
        FromUser = 'tsadmin'
        print(df_answers['name_y'].iloc[i])
        ToUser = df_answers['name_y'].iloc[i]#'tamofah%40CMC.LOCAL'
        print("Transfering Ownership for "+ df_answers['name_x'].iloc[i]+ " from "+ FromUser + " to " +ToUser)
        TransferOwnership(sess,FromUser,ToUser,ObjGUID)
    except:
        print("Failed to transfer Ownership for ")
        pass
```
